```python
import numpy as np
import matplotlib.pyplot as plt
#import seaborn
plt.rcParams['text.usetex'] = True
#plt.rcParams['text.latex.preamble'] = [r'\usepackage[cm]{sfmath}']
```

```python
xarray=np.linspace(-2.5,1.6,100)

y1 = lambda x: np.exp(x)
y2 = lambda x: x + 2
```

```python
fig,ax=plt.subplots(figsize=(9,5))

ax.plot(xarray,y1(xarray))
ax.plot(xarray,y2(xarray))

ax.axhline(y=0,color='black')
ax.axvline(x=0,color='black')

ax.text(0.4,0.37,r'$y=x+2$', transform=fig.transFigure, fontsize=20, rotation=19.5)
ax.text(0.65,0.35,r'$y=e^x$', transform=fig.transFigure, fontsize=20, rotation=25)
#ax.text(0.5,0.5,'asx', transform=fig.transFigure, fontsize=20)

ax.axis('off')
```

```python
fig.savefig('figures/intro_intersection.pdf')
```
